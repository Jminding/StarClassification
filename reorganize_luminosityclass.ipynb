```python
import pandas as pd
import numpy as np
import os
import re
```

```python
df = pd.read_csv("star_data_2.csv")
df
```

```python
star_sp = {}
stars = df["Star Name"].to_list()
sps = df["Spectral Class"].to_list()
def extract_info(input_string):
    # Extract the first capital letter
    capital_letter_match = re.search(r'[A-Z]', input_string)
    first_capital_letter = capital_letter_match.group() if capital_letter_match else None

    # Extract the first number (can be a float)
    number_match = re.search(r'[-+]?\d*\.\d+|\d+', input_string)
    first_number = float(number_match.group()) if number_match else None

    # Extract the first Roman numeral (I - V)
    roman_numerals = ["IV", "V", "III", "II", "I"]

    first_roman_numeral = None
    # Iterate through the string
    for i in range(len(input_string)):
        # Check for each Roman numeral
        for numeral in roman_numerals:
            if input_string[i:i+len(numeral)] == numeral:
                first_roman_numeral = numeral
                return first_capital_letter, first_number, first_roman_numeral
    return first_capital_letter, first_number, first_roman_numeral

for i in range(len(sps)):
    star_sp[stars[i]] = extract_info(sps[i])
star_sp
```

```python
luminosity_classes = [i[2] for i in star_sp.values()]
df["Luminosity Class"] = luminosity_classes
df.to_csv("star_data_3.csv")
df
```
