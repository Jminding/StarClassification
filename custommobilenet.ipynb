```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
mobile_net_weights_url = 'https://storage.googleapis.com/tfjs-models/tfjs/mobilenet_v1_1.0_224/model.json'
# download file using requests module
import requests
file = requests.get(mobile_net_weights_url)
```

```python
# save the file
with open('mobilenet_v1_1.0_224.json', 'wb') as f:
    f.write(file.content)
```

```python
# read the file
with open('mobilenet_v1_1.0_224.json', 'r') as f:
    mobilenet_json = f.read()
```

```python
# Load in the weights from json
import json
mobilenet_json_dict = json.loads(mobilenet_json)
# mobilenet_json_dict
mobilenet = tf.keras.models.model_from_json(mobilenet_json)
```
