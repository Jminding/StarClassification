```python
import albumentations as A
import cv2
import os
import matplotlib.pyplot as plt
```

```python
transform = A.Compose([
    A.HorizontalFlip(p=1),
    A.VerticalFlip(p=1),
])
```

```python
img = cv2.imread("/Users/jminding/Documents/Research/Star Images 2/O/Alnitak.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
transformed = transform(image=img)
transformed_image = transformed["image"]
plt.imshow(transformed_image)
```

```python
star_dir = "/Users/jminding/Documents/Research/Star Images 2"
for root, dirs, files in os.walk(star_dir, topdown=False):
    for name in files:
        if name.endswith(".jpg") and "_horizontalflipped" in name:
            print("Deleting " + name)
            os.remove(os.path.join(root, name))
```

```python
star_dir = "/Users/jminding/Documents/Research/Star Images 2"
for root, dirs, files in os.walk(star_dir, topdown=False):
    for name in files:
        if name.endswith(".jpg") and (("_horizontalflipped" not in name and "verticalflipped" not in name) or ("_90" in name or "_180" in name or "_270" in name)):
            print("Doing " + name)
            img = cv2.imread(os.path.join(root, name))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            transformed = transform(image=img)
            transformed_image = transformed["image"]
            cv2.imwrite(os.path.join(root, name.split(".")[0] + "_180.jpg"), cv2.cvtColor(transformed_image, cv2.COLOR_RGB2BGR))
```
